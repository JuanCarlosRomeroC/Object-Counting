## 2. Density maps visualization

```python
IMG_NAME = 'IMG_112'
IMG_FORMAT = 'jpg'
DATASET_SPLIT_PATH = '../../datasets/shanghai_tech/part_B/train'

IMG_PATH = f'{DATASET_SPLIT_PATH}/images/{IMG_NAME}.{IMG_FORMAT}'
DOTS_IMG_PATH = f'{DATASET_SPLIT_PATH}/gt_dots/{IMG_NAME}.png'
DENSITY_MAP_PATH = f'{DATASET_SPLIT_PATH}/gt_density_maps/{IMG_NAME}.npy'

SIGMA = config.SHANGHAI_TECH_PART_B_SIGMA
```

```python
img = skimage.io.imread(IMG_PATH)
dots_img = skimage.io.imread(DOTS_IMG_PATH) / 255.
density_map = np.load(DENSITY_MAP_PATH)

print('img_shape:', img.shape)
print('dots_img_shape:', dots_img.shape)
print('density_map_shape:', density_map.shape)
print('sigma for density_map_generation:', SIGMA)

plt.figure(figsize=(17, 12))
plt.subplot(1, 3, 1)
plt.title(IMG_NAME)
plt.imshow(img)

plt.subplot(1, 3, 2)
plt.title(f'Dots count: {dots_img.sum()}')
plt.imshow(dots_img, cmap='gray')

plt.subplot(1, 3, 3)
plt.title(f'Density map sum: {density_map.sum():.2f}')
plt.imshow(density_map, cmap='jet')
plt.colorbar(fraction=0.035, pad=0.04)
plt.show()
```
